```python
plt.figure(figsize=(20,15))

data.boxplot(column =[ 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat', 'medv'])
```

```python
q1 = data.quantile(0.25)
q3 = data.quantile(0.75)
iqr = q3 - q1
iqr

upper = q3 + (1.5 * iqr)
lower = q1 - (1.5* iqr)

data_out = data.drop(data[data['b'] >= upper['b']].index)
data_out = data.drop(data[data['b']<=lower['b']].index)
sns.boxplot(x='b',data = data)
plt.show()
sns.boxplot(x = 'b',data = data_out)
plt.show()
```
