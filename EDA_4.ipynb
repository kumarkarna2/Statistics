```python
import matplotlib.pylab as plt 
import pandas as pd
import seaborn as sns
import numpy as np
from scipy import stats

data = pd.read_excel('dataset/data.xlsx')
```

```python
data['Calories'].fillna(data['Calories'].mean(), inplace=True)
calData = data['Calories']
sns.displot(calData,kde=True, color='green')
plt.title('Before transformation')
# plt.show()

plt.scatter(calData, data['Maxpulse'])
plt.show()

```

```python
calData_sqrt = np.sqrt(calData)

sns.displot(calData_sqrt, kde=True,color='red')
plt.title('data after Square root transformation')
plt.show()
```

```python
calData_log = np.log1p(calData)
sns.displot(calData_log,kde=True, color='red')
plt.title('data after log transformation')
plt.show()
```

```python
calData_boxcox = stats.boxcox(calData)

sns.displot(calData_boxcox,kde =True, color='red')
plt.title('data after Boxcox transformation')
plt.show()
```

```python
calData_yeo = stats.yeojohnson(calData)
sns.displot(calData_yeo[0],kde=True, color='red')
plt.title('data after Yeo-Johnson transformation')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
import seaborn as sns

# import data from CSV file
data = pd.read_excel("dataset/data.xlsx")
A = data['Maxpulse']
B = data['Calories']

# create a figure with 3 rows and 3 columns
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(12,12))

# plot distribution plot before transformation in first column
axes[0, 0].set_title("Distribution Plot before Transformation")
sns.histplot(A, kde=True, ax=axes[0, 0])
sns.histplot(B, kde=True, ax=axes[1, 0])




```

```python
# apply square transformation and plot distribution plot and QQ plot in second column
A_square = np.square(A)
B_square = np.square(B)

axes[0, 1].set_title("Square Transformation")
sns.histplot(A_square, kde=True, ax=axes[0, 1])
sns.histplot(B_square, kde=True, ax=axes[1, 1])
stats.probplot(A_square, plot=axes[0, 2], dist='norm')
stats.probplot(B_square, plot=axes[1, 2], dist='norm')
```

```python
# apply log transformation and plot distribution plot and QQ plot in third column
A_log = np.log(A)
B_log = np.log(B)

axes[0, 1].set_title("Log Transformation")
sns.histplot(A_log, kde=True, ax=axes[0, 1])
sns.histplot(B_log, kde=True, ax=axes[1, 1])
stats.probplot(A_log, plot=axes[0, 2], dist='norm')
stats.probplot(B_log, plot=axes[1, 2], dist='norm')
```

```python
# apply box-cox transformation and plot distribution plot and QQ plot in fourth column
A_boxcox, _ = stats.boxcox(A)
B_boxcox, _ = stats.boxcox(B)

axes[0, 1].set_title("Box-Cox Transformation")
sns.histplot(A_boxcox, kde=True, ax=axes[0, 1])
sns.histplot(B_boxcox, kde=True, ax=axes[1, 1])
stats.probplot(A_boxcox, plot=axes[0, 2], dist='norm')
stats.probplot(B_boxcox, plot=axes[1, 2], dist='norm')

plt.tight_layout()
plt.show()
```
