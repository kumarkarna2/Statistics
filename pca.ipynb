```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.array([[90,60,90],[90,90,30],[60,60,60],[60,60,90],[30,30,30]])
for i in range(len(data)):
    print(data[i])
```

```python
mean = np.mean(data, axis=0)

rm = data - mean
print(mean)
```

```python
import numpy as np
import pandas as pd

#covariance matrix
cov = np.cov(data, rowvar=False, bias=True)
for i in range(len(cov)):
    print(cov[i])
```

```python
#eigan value and eigan vector
val,vect = np.linalg.eig(cov)
print("val = ",val)
print("vector :")
for i in range(len(vect)):
    print(vect[i])

#sort
i = np.argsort(val)[::-1]
val = val[i]
vect = vect[:,i]


print("\nafter sorting")
print("val = ",val)
print("vector :")
for i in range(len(vect)):
    print(vect[i])

```

```python
vectTran = vect[:,0:2]


result = np.dot(rm,vectTran)
for i in range(len(result)):
    print(result[i])
```
