```python
import matplotlib.pyplot as plt
```

```python
data = [[90,60,90],[90,90,30],[60,60,60],[60,60,90],[30,30,30]]
for i in range(len(data)):
    print(data[i])
```

```python
mean = np.mean(data, axis=0)

rm = data - mean
```

```python
import numpy as np
import pandas as pd

#covariance matrix
cov = np.cov(data, rowvar=False, bias=True)
for i in range(len(cov)):
    print(cov[i])
```

```python
#eigan value and eigan vector
val,vect = np.linalg.eig(cov)
print("val = ",val)
print("vector :")
for i in range(len(vect)):
    print(vect[i])

#sort
i = np.argsort(val)[::-1]
val = val[i]
vect = vect[:,i]


print("\nafter sorting")
print("val = ",val)
print("vector :")
for i in range(len(vect)):
    print(vect[i])

```

```python
vectTran = vect[:,0:2]


result = np.dot(rm,vectTran)
for i in range(len(result)):
    print(result[i])
```

```python
# pca
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pcaCom = pca.fit_transform(data)
pcaData = pd.DataFrame(data = data, columns = ['x', 'y', 'z'])


pca.explained_variance_
```

```python
pc = pca.transform(data)
plt.scatter(pc[:,0], pc[:,1], c='r')
```

```python
d3 = PCA(n_components=3)
# d3.fit(data)
d3_data = d3.fit_transform(data)
ax = plt.axes(projection='3d')
ax.scatter(d3_data[:,0], d3_data[:,1], d3_data[:,2], c='r')
```

```python
d3 = PCA(n_components=1)
d3.fit(data)
d3_data = d3.transform(data)
plt.scatter(d3_data[:,0], np.zeros(len(d3_data)), c='r')
```
